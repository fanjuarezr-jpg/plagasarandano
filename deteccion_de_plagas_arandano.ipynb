```python
# Cargar librerias
import os
import sys
import random
import requests
import numpy as np
import keras
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
import gdown
import zipfile
from ipywidgets import interact, widgets
from IPython.display import display
import matplotlib.pyplot as plt
import cv2
```

```python
def download_dataset_with_gdown(file_id, output_path):
    """
    Downloads a file from Google Drive using its file ID.
    
    Args:
        file_id (str): The Google Drive file ID.
        output_path (str): The local path to save the downloaded file.
    """
    try:
        # Construct the direct download URL format
        url = f'https://drive.google.com/uc?id={file_id}'
        
        # Download the file
        gdown.download(url, output_path, quiet=True)
        
        #print(f"\n Successfully downloaded file to: {output_path}")
        
    except Exception as e:
        print(f"\n An error occurred during download: {e}")
```

```python
def get_prediction(img):
    rimg = cv2.resize(img, (299, 299))
    rimg = np.expand_dims(rimg, axis=0)
    rimg = preprocess_input(rimg)
    rimg = rimg.astype('float32') / 255.
    res = inception_model.predict(rimg)
    idx_max = list(res[0])
    pred = categories[idx_max.index(max(idx_max))]
    return str(pred)
```

```python
# Upload an image to be processed by the DL model
uploader = widgets.FileUpload(
    accept='',
    multiple=False
)

label_1 = widgets.Label("INPUT: Imagen de hoja de arandano")
label_2 = widgets.Label("")
file = open(destination, "rb")
image = file.read()
ulimg = widgets.Image(
    value=image,
    format='png',
    width=300,
    height=400,
)

def on_upload_change(change):
    #label.value = "Loaded"
    uploaded_file = uploader.value
    value = uploaded_file[0].content.tobytes()
    ulimg.value = value
    nparr = np.frombuffer(value, np.uint8)
    cv_image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    label_2.value = "PREDICCION: " + get_prediction(cv_image)
    
uploader.observe(on_upload_change, names='value')
display(uploader, label_1, ulimg, label_2)
```
